```python
import time
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
df = pd.read_csv('HR_genres.csv')
df.head()
```

```python
start_time = time.time()
test = nx.Graph()
for node in range(1000):
    test.add_node(node)
time_spent = time.time() - start_time
print("Time spent : ", time_spent, "seconds")
```

```python
plt.figure(figsize=(20,20))
nx.draw(test)
```

```python
genres = df["genre"].unique().tolist()

random.seed(2018)
colors = []

for i in range(len(genres)):
    colors.append('%06X' % random.randint(0, 0xFFFFFF))

print(genres, colors)
genres_colors = {}
for i in range(len(genres)):
    genres_colors[genres[i]] = colors[i]
genres_colors
```

```python
plt.figure(figsize=(20,20))
G=nx.Graph()

G.add_nodes_from(range(10))

G.add_edges_from([(3, 4), (4, 5),(0,7),(2,6),(5,8),(9,7),(1,4),(2,8),(2,4),(0,1)], color='red')
pos=nx.fruchterman_reingold_layout(G) 

nx.draw_networkx_nodes(G,pos,
                       nodelist=range(10),
                       node_color=['#413D52', '#0E2D51', '#CEB98B', '#989AA0', '#3DA9B4', '#DE6ED7', '#78A745', '#788398', '#11E4B8', '#15085A', '#20C9D9', '#DE639D'],
                       node_size=[5000, 200, 2300, 1000,10000,300],
                       alpha=0.8)
nx.draw_networkx_edges(G,pos,[(3, 4), (4, 5),(0,7),(2,6),(5,8),(9,7),(1,4),(2,8),(2,4),(0,1)])
nx.draw_networkx_labels(G,pos,
                       font_size=20,
                       labels={4:'Pop', 1:'Indie Rock', 2:'Dance', 3:'International Pop', 0:'Rap/Hip Hop', 5:'Films/Games', 6:'Reggae', 7:'Folk', 8:'Indie Pop/Folk', 9:'Metal'}
                   )
plt.show()
```
